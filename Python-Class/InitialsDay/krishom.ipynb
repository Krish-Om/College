```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import plotly.express as ex
from sklearn.preprocessing import OneHotEncoder,StandardScaler,LabelEncoder
import os 
```

```python
fruits = {'apple':10,'mango':20,'papaya':30,'banana':12,'grapes':50,'watermelon':200,
          'orange':40,'litchi':80,'guava':33, 'kiwi':70,'avocado':90,'dragon fruit':700}

color =["cyan","magenta","black", "blue", "brown", "green", "orange", "pink", "red", "violet", "white", "yellow" ]
```

```python
size = fruits.values()
keylabel = fruits.keys()
# size,label
```

```python
plt.Figure(figsize=(30,15))
plt.pie(size,labels=keylabel,autopct='%0.01f%%',startangle=140,colors=color)
plt.title('This is the pie chart for the data science project')
plt.legend(loc='upper right',bbox_to_anchor=(2,1))
plt.tight_layout()
plt.savefig('piechart.png')
plt.show()
```
